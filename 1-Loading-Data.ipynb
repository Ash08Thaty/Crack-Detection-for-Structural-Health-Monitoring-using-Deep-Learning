## Import Libraries and Packages

```python
import os
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

## Load Images

```python
#reading negative files

negative_files = os.scandir('./Negative')
negative_files
```

```python
# grab the first file in the directory

file_name = next(negative_files)
file_name
```

```python
#reading the element if it is a file

os.path.isfile(file_name)
```

```python
# get image name

image_name = str(file_name).split("'")[1]
image_name
```

```python
# read image data

image_data = plt.imread('./Negative/{}'.format(image_name))
image_data
```

```python
# Dimension of image

print(image_data.shape)
print(
    "Image dimensions are {} pixels by {} pixels.".format(
        image_data.shape[0], image_data.shape[1]
    )
)
```

```python
plt.imshow(image_data)
```

```python
# looping through all the images in the *./Negative* directory & read and save them in the list **negative_images**

%%time

negative_images = []
for file_name in negative_files:
    if os.path.isfile(file_name):
        image_name = str(file_name).split("'")[1]
        image_data = plt.imread('./Negative/{}'.format(image_name))
        negative_images.append(image_data)
    
negative_images = np.array(negative_images)
```

```python
# saving the paths to the images in a variable instead of loading and saving the images

negative_images = os.listdir('./Negative')
negative_images
```

```python
negative_images.sort()
negative_images
```

```python
image_data = Image.open('./Negative/{}'.format(negative_images[0]))
```

```python
image_data
```

```python
plt.imshow(image_data)
```

```python
# looping through image directory and adding their save path

negative_images_dir = ['./Negative/{}'.format(image) for image in negative_images]
negative_images_dir
```

```python
# images w/o cracks

len(negative_images_dir)
```

```python
# Next 4 images

for i in range(1, 5):
    plt.imshow(Image.open(negative_images_dir[i]))
    plt.title("{} image".format(["2nd", "3rd", "4th", "5th"][i - 1]))
    plt.show()
```

```python
# Saving the paths to all the images in the *./Positive* directory in a list called **positive_images_dir**

positive_images_dir = ["./Positive/{}".format(filename) for filename in os.listdir("./Positive")]
positive_images_dir.sort()
positive_images_dir
```

```python
# images with cracks

print("There are {} images of cracked concrete.".format(len(positive_images_dir)))
```

```python
# first 4 images with crack

print("Cracked concrete:")
for i in range(4):
    plt.imshow(Image.open(positive_images_dir[i]))
    plt.title("{} image".format(["1st", "2nd", "3rd", "4th"][i]))
    plt.show()
```
